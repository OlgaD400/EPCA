```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import random
import pandas as pd
from EPCA.EPCA import EPCA
from helper_functions import create_noisy_datasets, run_epca
from plotting_functions import plot_compare_methods, plot_epca_trials
```

```python
from mnist import MNIST
mndata = MNIST('./data')
images, labels = mndata.load_training()
images = np.array(images)
labels = np.array(labels)
```

Choose an MNIST integer

```python
integer = 0
number = np.where(labels == integer)[0]
og_data = images[number, :]
n,m = og_data.shape
```

Get true PCs

```python
pca = PCA(n_components = 2)
pca.fit(og_data)
true_components = pca.components_
reconstruction_true = og_data@true_components.T@true_components

```

Create noisy data

```python
normal_white_data, uniform_white_data, sp_data, outlier_data = create_noisy_datasets(og_data, 
                                                                                    sp_probability = .01,
                                                                                    uniform_white_variance = 2, #max(np.average(data), 2) * 2,
                                                                                    normal_white_variance = 2, #max(np.average(data), 2) * 2,
                                                                                    outlier_scale = 10,
                                                                                    outlier_fraction = .05)
```

Generate EPCA PCs for a specified dataset.

```python
input_data = sp_data

epca = EPCA(num_components=2, num_bags=100, bag_size=100)
epca.run_epca(input_data)
indices, pred_components = epca.return_unique_vectors()
reconstruction_pred = og_data@pred_components.T@pred_components

```

Choose an image to project onto the PCs

```python
index = 100

plt.figure()
plt.imshow(reconstruction_true[index].reshape((28,28)))
plt.title('True Projection')
plt.show()

plt.figure()
plt.imshow(reconstruction_pred[index].reshape((28,28)))
plt.title('Predicted Projection')
plt.show()
```
