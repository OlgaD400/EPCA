```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
from EPCA import EPCA
from helper_functions import run_epca
from sklearn.cluster import DBSCAN
from varied_noise import outlier_comparison, white_noise_comparison, sparse_noise_comparison
from helper_functions import create_noisy_datasets
from plotting_functions import plot_epca_trials, plot_compare_methods
```

# Generate wave data

```python
def f(x,t):
    return (1 - 0.5 * np.cos(2 * t)) * (1 / np.cosh(x)) + (1 - 0.5 * np.sin(2 * t)) * (1 / np.cosh(x)) * np.tanh(x)
```

```python
n = 200
x = np.linspace(-10, 10, n)
t = np.linspace(0, 3000, 6000)

X,T = np.meshgrid(x,t)

data = f(X,T)
```

# Visualize True PCs and explained variance

```python
pca = PCA(n_components = 2)
pca.fit(data)

pca_1 = pca.components_[0,:]
pca_2 = pca.components_[1,:]

plt.figure()
plt.plot(pca.components_[0,:], label = 'pc1')
plt.plot(pca.components_[1,:], label = 'pc2')
plt.legend()

true_explained_variance = pca.explained_variance_ #explained_variance_ratio_
print('True explained variance', true_explained_variance)
# plt.plot(pca.components_[2,:])
```

# Create sample noisy data

```python
normal_white_data, uniform_white_data, sp_data, outlier_data = create_noisy_datasets(data, 
                                                                                    sp_probability = .01,
                                                                                    sparse_num = 0,
                                                                                    uniform_white_variance = 2,
                                                                                    normal_white_variance = 2,
                                                                                    outlier_scale = 10, 
                                                                                    outlier_fraction = .05)
```
