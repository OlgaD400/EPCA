# Outlier Varied Levels

```python
pca_1, pca_2, epca_1, epca_2, rpca_1, rpca_2 = load_comparison_data(
    root_filenames = ['iris_', 'wine_', 'breast_cancer_', 'wave_', 'mnist_0_', 
               'mnist_1_',])

fig, axs = plot_varied_levels_noise(
    first_component_performance=[pca_1, rpca_1, epca_1],
    second_component_performance=[pca_2, rpca_2, epca_2],
    feature_1=[2, 5, 10, 20, 100],
    feature_2=[1, 5, 10, 15, 20, 25],
    subfigure_title="Noise Scale",
    subfigure_caption="Noise Percentage",
    suptitle="Comparison of Outlier Noise Level",
    filepath="temp",
)
```

# White Noise Varied Levels

```python
pca_1, pca_2, epca_1, epca_2, rpca_1, rpca_2 = load_comparison_data(
    root_filenames=[
        "white_noise_iris_",
        "white_noise_wine_",
        "white_noise_breast_cancer_",
        "white_noise_wave_",
        "white_noise_mnist_0_",
        "white_noise_mnist_1_",
    ]
)

u_pca_1, u_pca_2, u_epca_1, u_epca_2, u_rpca_1, u_rpca_2 = load_comparison_data(
    root_filenames=[
        "u_white_noise_iris_",
        "u_white_noise_wine_",
        "u_white_noise_breast_cancer_",
        "u_white_noise_wave_",
        "u_white_noise_mnist_0_",
        "u_white_noise_mnist_1_",
    ]
)

pca_1 = np.hstack((pca_1, u_pca_1))
pca_2 = np.hstack((pca_2, u_pca_2))
rpca_1 = np.hstack((rpca_1, u_rpca_1))
rpca_2 = np.hstack((rpca_2, u_rpca_2))
epca_1 = np.hstack((epca_1, u_epca_1))
epca_2 = np.hstack((epca_2, u_epca_2))

fig, axs = plot_varied_levels_noise(
    first_component_performance=[pca_1, rpca_1, epca_1],
    second_component_performance=[pca_2, rpca_2, epca_2],
    feature_1=["Normal White Noise", "Uniform White Noise"],
    feature_2=[0.10, 1, 10, 100, 1000],
    subfigure_title=" ",
    subfigure_caption="Variance Divisor",
    suptitle="Comparison of White Noise Level",
    y1_step=30,
    y2_step=30,
    filepath="temp",
    figsize=(10, 10),
)
```

# Sparse Noise Varied Levels

```python
pca_1, pca_2, epca_1, epca_2, rpca_1, rpca_2 = load_comparison_data(
    root_filenames=[
        "sp_noise_iris_",
        "sp_noise_wine_",
        "sp_noise_breast_cancer_",
        "sp_noise_wave_",
        "sp_noise_mnist_0_",
        "sp_noise_mnist_1_",
    ]
)

fig, axs = plot_varied_levels_noise(
    first_component_performance=[pca_1, rpca_1, epca_1],
    second_component_performance=[pca_2, rpca_2, epca_2],
    feature_1=[0, 2, 5, 10],
    feature_2=[0.01, 0.05, 0.10],
    subfigure_title="Noise Scale:",
    subfigure_caption="Noise Probability",
    suptitle="Comparison of Sparse Noise Level",
    filepath="temp",
    y1_step = 30, 
    y2_step = 30
)

```
