```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from EPCA.EPCA import EPCA
from scipy.io import netcdf
import pandas as pd
from helper_functions import create_noisy_datasets
from plotting_functions import plot_epca_trials, plot_compare_methods
```

```python
# weekly SST data
sst_dict = netcdf.NetCDFFile('data/sst.wkmean.1990-present.nc', 'r').variables
# land-sea mask 
mask = netcdf.NetCDFFile('data/lsmask.nc', 'r').variables['mask'].data[0, :, :]
mask = mask.astype(float)
mask[mask == 0] = np.nan

sst_data = sst_dict['sst'].data.reshape((1726,-1))
```

Get true PCs.

```python
pca = PCA(n_components = 2)
pca.fit(sst_data)
true_components = pca.components_
reconstruction_true = sst_data@true_components.T@true_components
```

Create noisy data.

```python
normal_white_data, uniform_white_data, sp_data, outlier_data = create_noisy_datasets(sst_data, 
                                                                                    sp_probability = .01,
                                                                                    uniform_white_variance = 2, #max(np.average(data), 2) * 2,
                                                                                    normal_white_variance = 2, #max(np.average(data), 2) * 2,
                                                                                    outlier_scale = 10,
                                                                                    outlier_fraction = .05)
```

Generate EPCA PCs for a specified dataset.

```python
input_data = sp_data

epca = EPCA(num_components=2, num_bags=100, bag_size=20)
epca.run_epca(input_data)
indices, pred_components = epca.return_unique_vectors()
reconstruction_pred = input_data@pred_components.T@pred_components
```

Choose an image to project onto the PCs

```python
index = 100

plt.figure()
plt.imshow(reconstruction_true[index].reshape((180,360)))
plt.title('True Projection')
plt.show()

plt.figure()
plt.imshow(reconstruction_pred[index].reshape((180,360)))
plt.title('Predicted Projection')
plt.show()

```
